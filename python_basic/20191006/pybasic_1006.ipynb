## 만약 2위값, 3위값 등 한 값만 보고싶다면?

```python
rank_list = soup.select(".rank_list.v2")[0].select('.list_area') # class는 앞에 .을 붙인다.
rank_result_list = []
for rank in rank_list:
    rank_result = rank.text.strip().split("\n")
    rank_result_list += [{"rank": rank_result[0], "value": rank_result[1]}]

print(rank_result_list)
print("-" *100)

print('실시간 검색어 {}위는 "{}" 입니다.'.format(rank_result_list[0]["rank"], rank_result_list[0]["value"]))
```

```python
def find_rank_value(rank_num):
    rank_list = soup.select(".rank_list.v2")[0].select('.list_area') # class는 앞에 .을 붙인다.
    rank_result_list = []
    for rank in rank_list:
        rank_result = rank.text.strip().split('\n')
        rank_result_list += [{"rank": rank_result[0], "value": rank_result[1]}]
    print('실시간 검색어 {}위는 "{}" 입니다.'.format(rank_result_list[rank_num-1]["rank"], rank_result_list[rank_num-1]["value"]))
    
find_rank_value(14)
```

# Selenium

requests와 BeautifulSoup로는 작동하지 않는 까다로운 상황이 발생할 수 있다.

```python
from selenium import webdriver
from bs4 import BeautifulSoup

driver = webdriver.Chrome('/Users/Leila/Documents/GitHub/DataItGirls/python_basic/chromedriver') 
# 절대경로로 해야 문제없이돌아감

# 페이지가 모두 로드될 때까지 3초 기다려주는 코드
driver.implicitly_wait(3)
# 자바스크립트의 경우 읽는데 시간이 좀 걸림. 자바스크립트 외에도 여러 경우(인터넷이 느리다거나..) 읽는데 시간이 조금 필요하므로
# 약간 wait을 준다.

# url에 접근
driver.get('https://datalab.naver.com/keyword/realtimeList.naver')

# 소스를 읽어 BeautifulSoup를 이용해 파싱
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

rank_list = soup.select(".rank_list.v2")[0].select('.list_area')
rank_result_list = []
for rank in rank_list:
    rank_result = rank.text.strip().split('\n')
    rank_result_list += [{"rank": rank_result[0], "value": rank_result[1]}]
    
print(rank_result_list)
print("-"*100)
print('실시간 검색어 {}위는 "{}" 입니다.'.format(rank_result_list[0]["rank"], rank_result_list[0]["value"]))
```
